```python
import numpy as np
import matplotlib.pyplot as plt

# Funksjonen vi vil finne roten til: F(x) = 0
def F(x):
    return np.arctan(x) - 4/(1 + x*2)

# Derivert av F(x)
def F_derivert(x):
    return 1/(1 + x*2) - 4 * (-2*x) / (1 + x*2)**2

# Newtons metode
def newton(x0, tol=1e-12, max_iter=50):
    x = x0
    for i  in range(max_iter):
        x_new = x - F(x)/F_derivert(x)
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x

# Startverdi
x_start = 1.5

# Finn maksimumspunktet
x_max = newton(x_start)
y_max = np.exp(-x_max/4) * np.arctan(x_max)

print("x_max =", x_max)
print("f(x_max) =", y_max)

# Lag plottdata
xs = np.linspace(-5, 10, 1000)
ys = np.exp(-xs/4) * np.arctan(xs)

# Plot funksjonen
plt.plot(xs, ys, label="f(x) = e^{-x/4} arctan(x)")

# Marker toppunktet
plt.plot(x_max, y_max, 'ro', label=f"Max ({x_max:.4f}, {y_max:.4f})")

plt.title("Funksjonen f(x) og toppunkt funnet med Newtons metode")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
